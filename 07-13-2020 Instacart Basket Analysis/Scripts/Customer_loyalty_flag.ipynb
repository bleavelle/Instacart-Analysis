```python
# show avgs of users

op_merged.groupby('user_id').agg({'avg_order': ['max']})
```

booya! Next step: add a column that categorizes them as high spenders and low spenders <br> based on avg spending

```python
# new categories for spending averages as a for loop

result = []
for i in op_merged['avg_order']:
    if i < 10:
        result.append('Low spender')
    elif i >= 10:
        result.append('High spender')
    else:
        result.append("Insufficent Data")
```

```python
# save result to new column 'spender_category'

op_merged['spender_category'] = result
```
